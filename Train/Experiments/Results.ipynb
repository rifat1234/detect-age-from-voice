```python
import matplotlib.pyplot as plt

# x axis values
x = [5,10,25,50,75,100]
# corresponding y axis values
y = [0.602,0.723,0.795,0.845,0.876,0.891]

# x axis values
x1 = [5,10,25,50,75,100]
# corresponding y axis values
y1 = [0.518,0.67,0.767,0.820,0.859,0.882]

# plotting the points
plt.plot(x, y, color='green',
         marker='o', markerfacecolor='blue', markersize=9, label='KNN with gender')

plt.plot(x1, y1, color='green',
         marker='o', markerfacecolor='red', markersize=9, label='SVC with gender')

# setting x and y axis range
#plt.xlim(0,100)
plt.ylim(0.5,0.96)

# naming the x axis
plt.xlabel('Percentage of Data')
# naming the y axis
plt.ylabel('F1 Macro')

plt.legend(loc="lower right")

# giving a title to my graph
#plt.title('Some cool customizations!')
plt.xticks(x1)

# function to show the plot
plt.show()
```
